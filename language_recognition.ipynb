```python
import torch
from speechbrain.inference.classifiers import EncoderClassifier
from speechbrain.utils.fetching import LocalStrategy
import torch.nn.functional as F

# Load model
model = EncoderClassifier.from_hparams(
    source="speechbrain/lang-id-voxlingua107-ecapa",
    savedir="tmp",
    local_strategy=LocalStrategy.COPY
)

# Run classification
logits, max_logit, index, predicted_label = model.classify_file("audio_7285000Hz_14-30-45_13-01-2024.wav")

# Convert logits → probabilities
probs = F.softmax(logits, dim=1)

# Get top‑5 indices
top5_prob, top5_idx = torch.topk(probs, k=5)

# Convert to Python lists
top5_prob = top5_prob.squeeze().tolist()
top5_idx = top5_idx.squeeze().tolist()

# Map indices → labels
labels = model.hparams.label_encoder.decode_ndim(top5_idx)

# Print results
for p, lbl in zip(top5_prob, labels):
    print(f"{lbl}: {p:.4f}")
```
